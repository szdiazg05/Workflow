# Tokens y Costo

Consumo actual: 3.500 - 5.500 tokens por request
- Router: 500 tokens
- Retriever: 2.000 - 3.000 tokens (depende del contexto)
- Stylist: 1.000 tokens

Optimizaciones implementadas:
- Context engineering (solo 10 filas CSV, 4000 chars TXT)
- Reduccion 50x vs enviar archivos completos

Propuesta para escalar:
- RAG con embeddings
- Cache de respuestas frecuentes
- Batch processing para m√∫ltiples consultas

# Latencia

Mejoras propuestas:

- Usar modelos m√°s r√°pidos para routing
